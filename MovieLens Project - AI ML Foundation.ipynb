### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here
df_rating.head()
```

```python
df_rating['movie id'].value_counts()
```

```python
df_avg = df_rating[['movie id', 'rating']]
df_avg.head()
```

```python
df_avg1 = df_avg.groupby('movie id', as_index=False).count()
```

```python
df_avg1.shape
```

```python
df_avg1.head(258)
```

```python
df_sum = df_avg.groupby('movie id', as_index=False)['rating'].mean()
```

```python
df_sum.head()
```

```python
df_sum = df_sum.rename(columns={"rating": "avg rating"})
```

```python
df_sum.shape
```

```python
df_merge_avg_rating = pd.merge(df_avg1,df_sum,how='inner',on='movie id')
df_merge_avg_rating.head()
```

```python
df_merge_avg_rating.shape
```

```python
df_final_rating = df_merge_avg_rating[df_merge_avg_rating['rating']>100]
df_final_rating.head()
```

```python
df_final_rating.shape
```

```python
df_final_rating = df_final_rating.sort_values(by=['avg rating'], ascending=False)
df_final_rating.head()
```

```python
df_final_rating.shape
```

```python
df_final_rating_trim = df_final_rating.iloc[0:25,:]
df_final_rating_trim.shape
```

```python
df_top_movies = pd.merge(df_final_rating_trim,df_movie,how='inner',on='movie id')
df_top_movies.shape
```

```python
df_top_movies.head()
```
