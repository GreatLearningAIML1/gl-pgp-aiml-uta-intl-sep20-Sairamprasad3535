```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_reg = df.drop(columns = ['Target'])
X_reg.head()
```

```python
#Dealing with Outliers for Logistic Regression
Q1 = X_reg.quantile(0.25)
Q3 = X_reg.quantile(0.75)
#IQR = Q3 - Q1

Lower_Fence = Q1 - (1.5 * IQR)

Upper_Fence = Q3 + (1.5 * IQR)
print(Upper_Fence)
```

```python
for col in X_reg.columns:
    X_reg[col] = X_reg[col].clip(Lower_Fence[col], Upper_Fence[col])
```

```python
for col in ['job_admin.', 'job_blue-collar', 'job_entrepreneur',
       'job_housemaid', 'job_management', 'job_retired', 'job_self-employed',
       'job_services', 'job_student', 'job_technician', 'job_unemployed',
       'marital_divorced', 'marital_married', 'marital_single', 'default_no',
       'default_yes', 'housing_no', 'housing_yes', 'loan_no', 'loan_yes',
       'month_apr', 'month_aug', 'month_dec', 'month_feb', 'month_jan',
       'month_jul', 'month_jun', 'month_mar', 'month_may', 'month_nov',
       'month_oct', 'month_sep']:
    X_reg[col] = X_reg[col].astype('category')

X_reg.info()
```

```python
#Scalling the data using Min Max Scaller
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_reg[['age','education','balance','day','duration','campaign','pdays','previous']] = mms.fit_transform(X_reg[['age','education','balance','day','duration','campaign','pdays','previous']])
X_reg.head()
```

```python
x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, Y_reg, test_size = 0.3, random_state = 1)
```

```python
model = LogisticRegression()
model.fit(x_train_reg, y_train_reg)
```

```python
y_predict_reg = model.predict(x_test_reg)
```

```python
model_score = model.score(x_test_reg, y_test_reg)
print(model_score)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score,accuracy_score
```

```python
## function to get confusion matrix in a proper format
def draw_cm( actual, predicted ):
    cm = confusion_matrix( actual, predicted)
    sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = [0,1] , yticklabels = [0,1] )
    plt.ylabel('Observed')
    plt.xlabel('Predicted')
    plt.show()
```

```python
print("Trainig accuracy",model.score(x_train_reg,y_train_reg))  
print()
print("Testing accuracy",model.score(x_test_reg, y_test_reg))
print()
print('Confusion Matrix')
print(draw_cm(y_test_reg,y_predict_reg))
print()
print("Recall:",recall_score(y_test_reg,y_predict_reg))
print()
print("Precision:",precision_score(y_test_reg,y_predict_reg))
print()
print("F1 Score:",f1_score(y_test_reg,y_predict_reg))
print()
print("Roc Auc Score:",roc_auc_score(y_test_reg,y_predict_reg))
```
