```python
#Dealing with Outliers
Q1 = df_1.quantile(0.25)
Q3 = df_1.quantile(0.75)
IQR = Q3 - Q1

Lower_Fence = Q1 - (1.5 * IQR)

Upper_Fence = Q3 + (1.5 * IQR)
print(Upper_Fence)
```

```python
#Using Capping to negate outliers
for col in df_1.columns:
    df_1[col] = df_1[col].clip(Lower_Fence[col], Upper_Fence[col])
```

```python
#Checking if capping is done
df_1['slag'].max()
```

```python
#Checking distribution after capping
for col in df_1.columns:
    sns.distplot(df_1[col])
    plt.show()
```

**R2 Score:**
* R2 score is a metric used to guage how well the model has learnt the variations in the independent variables
* The score ranges between 0 to 1 with 0 menaning the model is not able to explain the variations and 1 being the model has learnt the variations well
* Hence higher the R2, better is the model
* In the current problem statement, Gradient Booseter model has the highest R2 and hence is the best model
* Gradient booster has learnt 90% of the variations in the independent variables

**Inference:**
* According to the importance in features of the best model, Age is the most important factor in determining the strength of concrete
* Second comes amount of cement followed by amount of water
* To get the best strength in concrete, the mixture must have greater age, cement and water
* Gradient boost model has the best accuracy in determining the strength of concrete with an accuracy of 90%
* 95% confidance interval of the model is 88% - 92% accuracy
* Outliers for 3 columns were treated
* Capping was used as the outliers were few in number and also were close to the maximum values
* Zeros in the dataset was not considered as missing values
